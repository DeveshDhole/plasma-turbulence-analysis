```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
import os

# --- File paths (choose one run) ---
Base_dir = "/home/UNN/w24021992/NAS/simulations_beta_scan/beta_p_4/beta_e_4"

File_Etot = f"{Base_dir}/EnTot_ApJ_.h5"
File_Time = f"{Base_dir}/Time_ApJ_.h5"

# --- Helper function ---
def load_dataset(fp):
    with h5py.File(fp, "r") as f:
        key = list(f.keys())[0]
        return f[key][:]

# --- Load data ---
E = load_dataset(File_Etot)
t = load_dataset(File_Time)

# --- Normalize ---
E_norm = E / E[0]

# --- Make plots directory if needed ---
plot_dir = f"{Base_dir}/plots"
os.makedirs(plot_dir, exist_ok=True)

# --- Plot ---
plt.figure(figsize=(8, 5))
plt.xlim(0, 300)
plt.ylim(0.98, 1.02)

plt.plot(t, E_norm)
plt.xlabel("Time")
plt.ylabel(r"$E / E_0$")
plt.title("Normalized Total Energy vs Time")
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig(f"{plot_dir}/EnergyPlot.png", dpi=300)
plt.show()
```
